# Problem 19.3

## Solution in Pyomo

```python
from pyomo.environ import *

model = ConcreteModel()
model.x1 = Var(initialize=0, bounds=(0,2))
model.x2 = Var(initialize=0, bounds=(0,2))

y = -0.1*model.x1**4 + 0.2*model.x2*model.x1**2 \
   - 0.09*model.x2**2 - 0.11*model.x1**2 + 0.15*model.x1 + 0.5

model.objective = Objective(expr = y, sense=maximize)

SolverFactory('ipopt').solve(model)
print('X1 =', model.x1())
print('X2 =', model.x2())
```

## Graphical Interpretation

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import numpy as np

x1 = np.linspace(0,2,100)
x2 = np.linspace(0,2,100)

x1_grid, x2_grid = np.meshgrid(x1,x2)

y = -0.1*x1_grid**4 + 0.2*x2_grid*x1_grid**2 \
    -0.09*x2_grid**2 - 0.11*x1_grid**2 + 0.15*x1_grid + 0.5

plt.figure(figsize=(10,10))
plt.contour(x1_grid, x2_grid, y, 100)
plt.plot(model.x1(), model.x2(), 'r.', ms=20)
plt.axis('square')
plt.grid()
```
