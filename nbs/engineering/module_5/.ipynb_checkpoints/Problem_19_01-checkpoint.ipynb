```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0,1,200)
tf = 5.3 * x * np.exp(-3.6*x + 2.7)
plt.plot(x, tf)
```

```python
from scipy.optimize import minimize_scalar

def func(x):
    return -(5.3 * x * np.exp(-3.6*x + 2.7))

minimize_scalar(func, method='brent', bracket=[0,1])
```
