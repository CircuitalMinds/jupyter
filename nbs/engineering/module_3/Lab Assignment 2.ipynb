The goals of Lab Assignment 2 is to familiarize you with other parts of the TCLab package, and to have you learn how to implement and test a simple proportional control loop.  For reference, here is a basic simulation of the TCLab response to a constant input on heater 1.

```python
from tclab import clock, setup, Historian, Plotter
TCLab = setup(connected=False, speedup=10)

MV = 40  # constant value to be applied to heater 1

with TCLab() as a:
    h = Historian(a.sources)
    p = Plotter(h)
    for t in clock(400):
        PV = a.T1         # get measurement measurement
        a.U1 = MV         # apply control action
        p.update()        # record results
```
