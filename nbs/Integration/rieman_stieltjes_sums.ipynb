# Riemann-Stieltjes Integral

```python
# import libraries
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.patches as patches
import matplotlib.animation as animation

# ignore warnings
import warnings
warnings.filterwarnings("ignore")

# set a random seed
np.random.seed(42)
```

## Implementation and testing

```python
def derivative(f, a, h=0.01):
    '''Approximates the derivative of the function f in a
    
    :param function f: function to differentiate
    :param float a: the point of differentiation
    :param float h: step size
    :return float: the derivative of f in a
    '''
    return (f(a + h) - f(a - h))/(2*h)
```

```python
def stieltjes_integral(f, g, a, b, n):
    '''Calculates the Riemann-Stieltjes integral based on the composite trapezoidal rule
    relying on the Riemann Sums.
    
    :param function f: integrand function
    :param function f: integrator function 
    :param int a: lower bound of the integral
    :param int b: upper bound of theintergal
    :param int n: number of trapezoids of equal width
    :return float: the integral of the function f between a and b
    '''
    eps = 1e-9
    h = (b - a)/(n + eps)  # width of the rectangle
    dg = lambda x: derivative(g, x, h=1e-8)  # derivative of the integrator function
    result = 0.5*f(a)*dg(a) + sum([f(a + i*h)*dg(a + i*h) for i in range(1, n)]) + 0.5*f(b)*dg(b)
    result *= h
    return result
```

```python
f = lambda x: x  # simple linear function
g = lambda x: 2 * x  # simple translation
a, b = 2, 5  # the intervals of integration
N = 300  # number of points to generate
T = b - a  # number of units
points_per_unit = N // T  # number of points per unit

t = np.linspace(0, b, points_per_unit*b + 1)
F = f(t)  # on the entire X-axis

X = np.linspace(a, b, N + 1)  # generate more points on the x-axis
Y = f(X)  # original function
G = g(X)  # transformation
```

```python
y_int = stieltjes_integral(f, g, a, b, 100)  # calculate the Stieltjes integral
err = np.abs(21.0 - y_int)  # calculate the error
print('The calculated value of the integral is {:.6f}, the error is {:g}'.format(y_int, err))
```

```python
y_int = stieltjes_integral(lambda x: x, lambda x: x*x, 0, 1, 1000)  # calculate the Stieltjes integral
err = np.abs(2/3 - y_int)  # calculate the error
print('The calculated value of the integral is {:.6f}, the error is {:g}'.format(y_int, err))
```

# Riemann-Stieltjes Illustration

```python
f1 = lambda x: x
f2 = lambda x: 1.0 * x / 3.
g = lambda x: 3.0 * x

A, B = 0, 16  # intervals for the full function
N = 1600  # number of points to generate
pt_unit = (N / (B - A))  # num of points per unit

X = np.linspace(A, B, N + 1)  # generate points on the x-axis
Y1 = f1(X)  # original function
Y2 = f2(X)  # scaled function

a1, b1 = 2, 5  # intervals for the original function
idx_a1, idx_b1 = int(a1 * pt_unit), int(b1 * pt_unit)

a2, b2 = 6, 15  # intervals for the scaled function
idx_a2, idx_b2 = int(a2 * pt_unit), int(b2 * pt_unit)


# bar plot for the first function
num_trapezoids1 = 5
x1 = np.linspace(a1, b1, num_trapezoids1 + 1)
y1 = f1(x1)

x_mid1 = (x1[:-1] + x1[1:])/2
y_mid1 = f1(x_mid1)

# bar plot for the second function
num_trapezoids2 = 5
x2 = np.linspace(a2, b2, num_trapezoids2 + 1)
y2 = f2(x2)

x_mid2 = (x2[:-1] + x2[1:])/2
y_mid2 = f2(x_mid2)
```

```python
fig = plt.figure(figsize=(21, 10))  # make a new figure

ax = plt.axes(xlim=(A, B), ylim=(0, B))
ax.set_xticks(np.linspace(A, B, (B - A) + 1))
ax.tick_params(labelsize=18)

ax.plot(X, Y1, color='#2CBDFE', lw=2)
ax.plot(X[idx_a1:idx_b1], Y1[idx_a1:idx_b1], color='#2CBDFE', lw=5)
ax.plot(X[idx_a2:idx_b2], Y2[idx_a2:idx_b2], color='#2CBDFE', lw=5)

ax.plot([a1], [Y1[idx_a1]], marker='.', markersize=20, c='k')  # black point at (a1, f1(a1))
ax.plot([b1], [Y1[idx_b1]], marker='.', markersize=20, c='k')  # black point at (b1, f1(b1))

ax.plot([a2], [Y2[idx_a2]], marker='.', markersize=20, c='r')  # black point at (a2, f2(a2))
ax.plot([b2], [Y2[idx_b2]], marker='.', markersize=20, c='r')  # black point at (b2, f2(b2))

ax.text(a2, Y1[idx_a1] + 0.5, '$f(a)$', fontsize=30)  # text to point (a2, f(a1))
ax.text(b2, Y1[idx_b1] + 0.3, '$f(b)$', fontsize=30)  # text to point (b2, f(b1))

# horizontal line (a1, f1(a1)) -> (a2, f2(a2))
ax.plot([a1, a2 - 0.18], [Y1[idx_a1], Y2[idx_a2]], marker='>', ms=15, markevery=[False, True], ls='-.', lw=2, c='#9D2EC5')
# horizontal line (b, f(b)) -> (b+3, f(b))
ax.plot([b1, b2 - 0.18], [Y1[idx_b1], Y2[idx_b2]], marker='>', ms=15, markevery=[False, True], ls='-.', lw=2, c='#9D2EC5')

ax.bar(x_mid1, y_mid1, width=1.0*(b1-a1)/num_trapezoids1, alpha=0.1, align='center', color='b', edgecolor='b', lw=2)
ax.bar(x_mid2, y_mid2, width=1.0*(b2-a2)/num_trapezoids2, alpha=0.4, align='center', color='#47DBCD', edgecolor='#47DBCD', lw=2)

plt.savefig('riemann_stieltjes_illustrated.png', dpi=300, bbox_inches='tight')
```
