# TCLab Simulation

```python
import simpy
import sys
sys.path.append("../TCLab")
from TCLab import TCLab


def model(env):
    global T1
    Tamb = 20
    Th = Tamb
    Ts = Tamb
    dt = 1
    while True:
        T1 = Ts
        yield env.timeout(dt)
        dTh = 0.01*Q1 + 0.02*(Tamb - Th)
        dTs = 0.01*(Th - Ts) + 0.01*(Tamb - Ts)
        Th += dt*dTh
        Ts += dt*dTs  


def heaterInput(env):
    global Q1
    tperiod = 200
    amplitude = 25
    mean = 75
    while True:
        Q1 = mean + amplitude
        amplitude = -amplitude
        yield env.timeout(tperiod/2)


def reporter(env):
    while True:
        print(env.now, Q1, T1)
        yield env.timeout(5)


env = simpy.Environment()
env.process(heaterInput(env))
env.process(model(env))
env.process(reporter(env))
env.run(until=500)
```
