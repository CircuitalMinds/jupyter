```python
import numpy as np
from PIL import Image
from scipy import ndimage
```

```python
src = np.array(Image.open('data/src/lena.jpg').convert('L'))
```

```python
print(src.shape)
```

```python
k_33 = np.ones((3, 3)) / 9
dst_33 = ndimage.convolve(src, k_33)
dst_33_im = Image.fromarray(np.uint8(dst_33))
```

```python
dst_33_im
```

```python
k_55 = np.ones((5, 5)) / 25
dst_55 = ndimage.convolve(src, k_55)
dst_55_im = Image.fromarray(np.uint8(dst_55))
```
